```python
import csv
import uuid
import secrets
import string

# CONFIG
DOMAIN = "cazehiresense.com"
TOTAL_USERS = 60
EMAIL_LENGTH = 7
QUOTA_MB = 0
CSV_FILE = "mailboxes_dev.csv"

# character sets
EMAIL_CHARS = string.ascii_lowercase + string.digits
PASSWORD_CHARS = string.ascii_letters + string.digits + "!@#$%^&*()"

def generate_email_local():
    return 'hs' + ''.join(secrets.choice(EMAIL_CHARS) for _ in range(EMAIL_LENGTH))

def generate_password(uid):
    random_part = ''.join(secrets.choice(PASSWORD_CHARS) for _ in range(12))
    return f"{random_part}!{uid[:8]}"

with open(CSV_FILE, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    
    # header
    writer.writerow(["email", "password", "name", "quota", "app_password"])

    for i in range(1, TOTAL_USERS + 1):
        uid = uuid.uuid4().hex
        local_part = generate_email_local()
        email = f"{local_part}@{DOMAIN}"
        password = generate_password(uid)
        name = f"User {i}"

        writer.writerow([email, password, name, QUOTA_MB])

print(f"✅ {TOTAL_USERS} mailboxes written to {CSV_FILE}")
```
